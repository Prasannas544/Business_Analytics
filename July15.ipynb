```r
X <- mtcars$wt
Y <- mtcars$mpg
```

```r
plot(X,Y,main="Mileage vs weight",xlab="Weight",ylab="Mileage",pch=17,frame=FALSE) # Change point shape (pch = 19) and remove frame.
```

```r
#Add regression line
plot(X,Y,main="Mileage vs weight",xlab="Weight",ylab="Mileage",pch=17,frame=FALSE)
abline(lm(Y~X,data= mtcars),col="green")
```

```r
#Compute the density data
dens <- density(mtcars$mpg)
# plot density
plot(dens,frame=FALSE,main="Density plot of mpg")
polygon(dens,col="yellow")
```
